```python
random.seed(13)
np.random.seed(13)

t = np.linspace(0, 1, 1024)

#линейная частота сезона
f = 10 #P = 1/10

S1 = np.sin(2*np.pi*2*f*t)
S2 = np.sin(2*np.pi*4*f*t)
S3 = np.sin(2*np.pi*8*f*t)
S4 = np.sin(2*np.pi*16*f*t)

F = S1 + S2 + S3 + S4

N = 1024

plt.figure(figsize = (15, 5))
plt.plot(t, F, 'k', label = 'ts')
plt.plot(t, S1, 'b', label = 'S1')
plt.plot(t, S2, 'r', label = 'S2')
plt.plot(t, S3, 'g', label = 'S3')
plt.plot(t, S4, 'g', label = 'S4')
plt.legend()
plt.show()
```

```python
def SSA_modes(F, L):
    N = len(F)
    K = N - L + 1
    X = np.empty((L, K))
    #шаг вложения
    for i in range(0, L): #rows
        for j in range(0, K):
            X[i,j] = F[i+j]
    #X = np.column_stack([F[i:i+L] for i in range(0,K)])
    #шаг сингулярного разложени]
    S = np.dot(X, X.T)
    U, A, _ = np.linalg.svd(S)
    # U - матрица собственных векторов
    # A - массив собственных чисел (в неубывающем порядке)
    V = np.dot(X.T, U) #матрица траекторных векторов
    return A, U, V
def SSA_group(A, U, V, N, I):
    
    L = len(A)
    K = N - L + 1

    #шаг группировки
    V = V.transpose()
    Z = np.dot(U[:, I], V[I, :])

    #этап диагонального усреднения
    G = np.zeros(N)
    L_ = min(L, K)
    K_ = max(L, K)    
    
    for k in range(0, L_):
        for m in range(0, k + 1):
            G[k] += Z[m][k-m]
        G[k] *= 1/(k+1)
    for k in range(L_, K_):
        for m in range(0, L_):    
            G[k] += Z[m][k-m]   
        G[k] *= 1/L_
    for k in range(K_, N):
        for m in range(k-K_+1, N-K_+1):
            G[k] += Z[m][k-m]
        G[k] *= 1/(N-k)
    return G
```

## SSA-R

```python
def SSA_R(F, N, L, M, I):
    A, U, V = SSA_modes(F, L)
    vu = np.linalg.norm(U[-1, I])
    R = np.sum(U[L - 1, I] * U[0:L - 1, I], 1)
    R = R / (1 - vu * vu)
    Q = F 
    for i in range(N, N + M): 
        ag = [] 
        for j in range(0, L-2): 
            ag.append(R[j] * Q[i-j-1]) 
        Q = np.append(Q, sum(ag))
    return Q
```

## SSA-S

```python
def SSA_S(F, N, L, M, I):
    for f_new in range(M):
        #print(f_new)
        F = np.append(F, random.uniform(min(F), max(F)))
        while True:
            
            N = len(F)
            # шаг вложения + шаг сингулярного разложения
            A, U, V = SSA_modes(Q, L) 
            L = len(A)
            K = N - L + 1

            #шаг группировки
            V = V.transpose()
            Z = np.dot(U[:, I], V[I, :])

            #этап диагонального усреднения
            G = np.zeros(N)
            L_ = min(L, K)
            K_ = max(L, K)    
    
            for k in range(0, L_):
                for m in range(0, k + 1):
                    G[k] += Z[m][k-m]
                G[k] *= 1/(k+1)
            for k in range(L_, K_):
                for m in range(0, L_):    
                    G[k] += Z[m][k-m]   
                G[k] *= 1/L_
            for k in range(K_, N):
                for m in range(k-K_+1, N-K_+1):
                    G[k] += Z[m][k-m]
                G[k] *= 1/(N-k)

            
            
            if round(F[-1],5) - round(G[-1],5) == 0:
                break
            else:
                F[-1] = G[-1]
    return F
```
